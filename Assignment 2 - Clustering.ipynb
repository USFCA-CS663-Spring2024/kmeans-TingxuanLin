  * A completeness score of 0.99 indicates that almost all instances of the same class are in the same cluster, suggesting high completeness.
3.
  * Homogeneity score measures the homogeneity of the clustering with respect to the true labels. It evaluates whether each cluster contains only members of a single class.
  * A homogeneity score of 0.99 suggests that each cluster contains instances from a single class, indicating high homogeneity.
4.
  * V-Measure score is the harmonic mean of homogeneity and completeness. It provides a balance between these two measures.
  * A V-measure score of 0.99 indicates high agreement between the clustering and the true labels, considering both homogeneity and completeness.

* There are null value in df, I choose to drop na value, I think deleting na value of position information will not affect the final results.

```python
import pandas as pd

df = pd.read_csv('Taxi_Trips.csv')
df
```

```python
pick_up = df.loc[:, ["Pickup Centroid Latitude", "Pickup Centroid Longitude"]].dropna()
pick_up.columns = ["pLat", "pLon"]
pick_up = pd.DataFrame(pick_up, columns=["pLat", "pLon"])
pick_up
```

### Choosing KMeans to do clustering
* dataset is too big, KMEANS is the most efficient
* I can adjust the number of clusters by inertia to find the most fit clustering
* Based on inertia plotting in range (2, 19), I find the elbow value might be 5
* The result is what I expected, because we can see from plot. The most dense cluster is '0' and '3'. '0' is near north side in Chicago. That area belongs to business center, CBD. '3' is the north side of Chicago. Most population live in this part, the house buiding is dense.
* In my case, the silhouette coefficient is about 0.62, which indicates that the clustering results are relatively reasonable. Specifically, most samples are densely clustered within their clusters and are clearly separated from other clusters.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(pick_up)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
import seaborn as sns

sns.scatterplot(data=pick_up, x=pick_up.pLon, y=pick_up.pLat, hue=kmeans.labels_, 
                s=20, alpha=1)
```

```python
kmeans.inertia_
```

```python
kminertia = pd.DataFrame(data=[], index=range(2, 19), columns=['inertia'])

for k in range(2, 19):
    kmeans= KMeans(n_clusters=k)
    kmeans.fit(pick_up)
    kminertia.loc[k] = kmeans.inertia_
```

```python
kminertia
```

```python
kminertia.plot()
```

```python
from sklearn.metrics import silhouette_score

ss = silhouette_score(pick_up, kmeans.labels_)
print("silhouette score: %s" % ss)
```

## Implementing DBSCAN to analyze mopsi_data
* I choose DBSCAN, firstly, the size of data is not too big. Second, there is no na value in the dataset, so nothing will affect the dataset. Then, I do not know the number of clusters in advance, DBSCAN can find out the shapes and sizes automatically.
* Using "Grid Searsh" approach to find better eps and min_samples
* In my case, the silhouette coefficient is about 0.64, which indicates that the clustering results are relatively reasonable. Specifically, most samples are densely clustered within their clusters and are clearly separated from other clusters.

```python
from sklearn.model_selection import ParameterGrid
# Try different values
param_grid = {
    'max_dist': [0.1, 0.2, 0.3], 
    'hoodsize': [3, 5, 7]  
}

best_score = -1
best_params = {}

for params in ParameterGrid(param_grid):
    max_dist = params['max_dist']
    hoodsize = params['hoodsize']
    
    cluster = DBSCAN(eps=max_dist, min_samples=hoodsize)
    
    cluster_labels = cluster.fit_predict(df)
    
    score = silhouette_score(df, cluster_labels)
    
    if score > best_score:
        best_score = score
        best_params = params

print("Best parameters:", best_params)
print("Best silhouette score:", best_score)
```

```python
from sklearn.cluster import DBSCAN
max_dist = 0.1
hoodsize=7
cluster = DBSCAN(eps=max_dist, min_samples=hoodsize)
cluster.fit(df)
# Association between observations & clusters.
print(cluster.labels_)
```

```python
sns.scatterplot(data=df, x=df["lon"], y=df["lat"],
                hue=cluster.labels_, alpha=0.4)
```

```python
from sklearn.metrics import silhouette_score

ss = silhouette_score(df, cluster.labels_)

print("Silhouette Coefficient: %s" %  ss)
```
